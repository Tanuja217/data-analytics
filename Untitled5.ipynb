```python
# prompt: generate a code for drop down also include piechart

import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd
!pip install dash

df = pd.read_csv('/content/country_wise_latest.csv')
df = df.drop('Confirmed', axis=1)
df = df.drop(['Deaths', 'Recovered'], axis=1)


app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Country-wise Analysis"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in df['Country/Region'].unique()],
        value=df['Country/Region'].unique()[0]  # Set initial value
    ),
    dcc.Graph(id='pie-chart')
])


@app.callback(
    Output('pie-chart', 'figure'),
    [Input('country-dropdown', 'value')]
)
def update_pie_chart(selected_country):
    country_df = df[df['Country/Region'] == selected_country]
    if not country_df.empty:
        # Calculate active cases (assuming 'Active' column represents active cases)
        active = country_df['Active'].values[0]

        # Create data for pie chart with 'Active' and 'Not Active' (as a placeholder)
        pie_data = pd.DataFrame({'Status': ['Active', 'Not Active'], 'Values': [active, 100-active]})
        fig = px.pie(pie_data, values='Values', names='Status', title=f'Active Cases in {selected_country}')
        return fig
    else:
        # Return an empty figure if there is no data for the selected country
        return px.pie()


if __name__ == '__main__':
    app.run_server(debug=True)
```
